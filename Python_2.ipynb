Para poder usar dataframe vamos a activar dos bibliotecas, una de ellas es numpy y la otra es pandas, las dos se complementan, y hacen cosas diferentes, para crear dataframe existen varias posibilidades, sin embargo, la mas usada es la lectura directa desde un archivo.

Opción 1

```python
import pandas as pd #Importo pandas y le pongo el sobrenombre pd
import numpy as np #Importo numpy y le pongo el sobrenombre np
```

```python
#La primera forma de crear un dateframe es usando el metodo array, y luego guardarlo en una variable
dataframe_1=np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
#Para crear el dataframe del arreglo anterior debemos usar pandas
pd.DataFrame(dataframe_1)
```

```python
#Ejercicio
dataframe_2=np.array([[5,10,15],[20,25,30],[35,40,45],[50,55,60]])
```

```python
pd.DataFrame(dataframe_2)
```

```python
#Ejercicio 2
dataframe_3=np.array([[3,6,9],[12,15,18],[21,24,27],[30,33,36]])
```

```python
pd.DataFrame(dataframe_3)
```

```python
#Ahora vamos a cambiarle el nombre a las columnas, esto se hace de la siguiente manera
dataframe_1=pd.DataFrame(dataframe_1, index=["fila1","fila2","fila3","fila4"],
                         columns=["columna1","columna2"])
```

Opción 3

Para este metodo uso los diccionarios de la siguiente manera

```python
#Primero creamos dos listas, puedo crear cuantas listas quiera, entendiendo que las listas son las columnas del dataframe
ciudadesColombia=["Bogotá","Medellín","Cali","Barranquilla"]
ciudadesInfo=[11000000,2500000,2200000,1300000]
```

```python
#Ahora creamos los diccionarios con las listas
ciudades={"ciudades":ciudadesColombia, "población":ciudadesInfo}
```

```python
#Le paso el metodo de pandas
dataframe_3=pd.DataFrame(ciudades)
```

```python
#Llamo el dataframe
dataframe_3
```

```python
#Ejercicio 4: 4 paises, población y expectativa de vida
paisesMundo=["Holanda","Australia","Ecuador","Brasil"]
paisesInfo=[17700000,26000000,18000000,215000000]
paisesExpec=[82,82,80,76]
```

```python
paises={"Paises":paisesMundo, "población":paisesInfo ,"expectativa":paisesExpec}
```

```python
dataframe_3=pd.DataFrame(paises)
```
