```python
#Ahora vamos a calcular el promedio de los datos de estas personas
(df_estudiantes['Total'])/3
```

```python
df_estudiantes
```

```python
#IMPORTANTE: Para descargar datos
df_estudiantes.to_csv('df_estudiantes.csv', index=False)
```

```python
#Ahora vamos a organizar de forma decendente
df_estudiantes.sort_values('math score', ascending=False)
```

```python
#También podemos ordenar los dataframe no solo por una columna sino por dos
#Hay que tener en cuenta, que organizará por el primer dato y el segundo será dependiente
df_estudiantes.sort_values(['reading score', 'writing score'], ascending=False)
```

```python
#De esta manera puedo guardar los datos, y sobre escribir, sin re usar la variable, con el metodo inplace
organizado_por_lectura=df_estudiantes.sort_values(['reading score', 'writing score'], ascending=False,
                                                  inplace=True)
```

```python
df_estudiantes
```

```python
organizado_por_lectura
```

```python
organizado_por_lectura.to_csv('organizado_por_lectura.csv', index=False)
```

```python
#Vamos a ordenar los valores por texto, para esto usaremos la función lambda
#Con ella convertimos primero los valores en minúsculas y luego ordenamos
df_estudiantes.sort_values('race/ethnicity', ascending=True,
                           key=lambda col:col.str.lower())
```
